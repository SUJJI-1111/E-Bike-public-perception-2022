# ðŸš² E-Bike Market Research Analysis
This notebook analyzes survey insights from a study on **E-Bikes before and after accidents**. It produces key findings, visualizations, and statistical tests from my research survey.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

sns.set(style='whitegrid')
```

```python
# Awareness of E-Bikes
awareness_data = pd.DataFrame({
    "Response": ["Yes", "No"],
    "Percentage": [94, 6]
})

plt.pie(awareness_data["Percentage"], labels=awareness_data["Response"], autopct="%1.0f%%", colors=["#66b3ff","#ff9999"])
plt.title("Awareness of E-Bikes")
plt.show()
```

```python
# Awareness of accidents
accident_data = pd.DataFrame({
    "Response": ["Yes", "No"],
    "Percentage": [76, 24]
})

plt.pie(accident_data["Percentage"], labels=accident_data["Response"], autopct="%1.0f%%", colors=["#99ff99","#ffcc99"])
plt.title("Awareness of Accidents")
plt.show()
```

```python
# Future intention to use
future_data = pd.DataFrame({
    "Response": ["Yes", "No"],
    "Percentage": [14, 86]
})

plt.pie(future_data["Percentage"], labels=future_data["Response"], autopct="%1.0f%%", colors=["#ffcc99","#c2c2f0"])
plt.title("Future Intention to Use E-Bikes")
plt.show()
```

```python
# Impression of E-Bikes after accidents
impression_data = pd.DataFrame({
    "Response": ["Good", "Bad"],
    "Percentage": [42, 58]
})

plt.pie(impression_data["Percentage"], labels=impression_data["Response"], autopct="%1.0f%%", colors=["#66b3ff","#ff9999"])
plt.title("Impression of E-Bikes After Accidents")
plt.show()
```

```python
# Chi-Square Test: Occupation vs Willingness to Use
# Example contingency table (dummy counts, based on report structure)
data = [[40, 60],   # Working Professionals: Yes, No
        [20, 80]]   # Students: Yes, No

chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

if p < 0.05:
    print("âœ… Reject Null Hypothesis: Occupation has an impact on willingness to use E-Bikes.")
else:
    print("âŒ Fail to Reject Null Hypothesis: No significant impact of occupation.")
```

## ðŸ” Observations
- **94%** of respondents are aware of E-Bikes.  
- **76%** are aware of accidents involving E-Bikes.  
- Only **14%** intend to use them in the future â†’ strong reluctance.  
- **58%** have a negative impression after hearing about accidents.  
- Chi-square test suggests occupation influences willingness to use.  

## âœ… Outcome
This analysis highlights that while awareness of E-Bikes is high, adoption is hindered by **safety concerns**.  
Accidents have created a negative perception, reducing willingness to adopt, especially among certain occupation groups.
